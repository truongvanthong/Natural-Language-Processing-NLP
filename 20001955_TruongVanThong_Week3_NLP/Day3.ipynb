### Xác suất (probability)

$$probability = \frac{event(s)}{outcome(s)}$$


### Xác suất có điều kiện (conditional probability)
$$P(A|B) = \frac{P(A \cap B)}{P(B)}$$

### Bayes Formula
$$P(Y|X) = \frac{P(X|Y)P(Y)}{P(X)}$$

$$Posterior = \frac{Likelihood \times Prior}{Evidence}$$

#### Rule:
- Product Rule: $P(A \cap B) = P(A|B)P(B) = P(B|A)P(A)$
- Sum Rule: $P(A \cup B) = P(A) + P(B) - P(A \cap B)$
- Theory of Total Probability: $P(A) = \sum_{i=1}^{n} P(A|B_i)P(B_i)$

Bài 1: Có 30 đề thi trong đó có 10 đề khó, 20 đề trung bình. Tìm xác suất để:

a. Một học sinh bắt một đề gặp được đề trung bình <br>
b. Một học sinh bắt hai đề, được ít nhất một đề trung bình <br>

Giải:

Câu a: Vì có 30 đề thi trong đó có 20 đề trung bình nên xác suất để một học sinh bắt một đề gặp được đề trung bình là: <br>
$$P(A) = \frac{Số đề TB}{Tổng số đề}= \frac{20}{30} = \frac{2}{3}$$

Câu b: 
- Xác suất lấy 2 đề khó:
    - Lấy đề khó đầu tiên: $\frac{10}{30}$
    - Sau khi lấy một đề khó, còn 9 đề khó và 20 đề trung bình, xác suất lấy được đề trung bình là: $\frac{9}{29}$
    - Xác suất lấy 2 đề khó là: $\frac{10}{30} \times \frac{9}{29} = \frac{3}{29}$

- Xác suất lấy ít nhất 1 đề dễ:
Để tính xác suất lấy ít nhất 1 đề dễ khi lấy 2 đề, ta trừ xác suất lấy 2 đề khó ra khỏi 1:
$$P(A) = 1 - P(\overline{A}) = 1 - \frac{3}{29} = \frac{26}{29}$$

Câu 2: Lớp có 100 Sinh viên, trong đó có 50 SV giỏi Anh Văn, 45 SV giỏi Pháp Văn, 10 SV giỏi cả hai ngoại ngữ. Chọn ngẫu nhiên một sinh viên trong lớp. Tính xác suất:

a) Sinh viên này giỏi ít nhất một ngoại ngữ.

b) Sinh viên này không giỏi ngoại ngữ nào hết.

c) Sinh viên này chỉ giỏi đúng một ngoại ngữ.

d) Sinh viên này chỉ giỏi duy nhất môn Anh Văn.

Giải:

a) Xác suất để sinh viên này giỏi ít nhất một ngoại ngữ là: <br>
$$P(A) = \frac{Số SV giỏi ít nhất 1 ngoại ngữ}{Tổng số SV} = \frac{50 + 45 - 10}{100} = \frac{85}{100} = 0.85$$

b) Xác suất để sinh viên này không giỏi ngoại ngữ nào hết là: <br>
$$P(B) = \frac{Số SV không giỏi ngoại ngữ nào hết}{Tổng số SV} = 1-0.85 = 0.15$$

c) Xác suất để sinh viên này chỉ giỏi đúng một ngoại ngữ là: <br>
$$P(C) = \frac{Số SV chỉ giỏi đúng 1 ngoại ngữ}{Tổng số SV} = \frac{50 + 45 - 10 - 10}{100} = \frac{75}{100} = 0.75$$

d) Xác suất để sinh viên này chỉ giỏi duy nhất môn Anh Văn là: <br>
$$P(D) = \frac{Số SV chỉ giỏi Anh Văn}{Tổng số SV} = \frac{50 - 10}{100} = \frac{40}{100} = 0.4$$

Câu 3: Trong một hộp có 12 bóng đèn, trong đó có 3 bóng hỏng. Lấy ngẫu nhiên không hoàn lại ba bóng để dùng. Tính xác suất đê:

a) Cả ba bóng đều hỏng.

b) Cả ba bóng đều không hỏng?

c) Có ít nhất một bóng không hỏng?

d) Chỉ có bóng thứ hai hỏng?

Giải:
- Tổng số cách chọn 3 bóng: $C_{12}^3 = 220$ <br>
    a) Xác suất để cả ba bóng đều hỏng là: <br>
    - Số cách chọn 3 bóng hỏng: $C_3^3 = 1$
    $$P(A) = \frac{Số cách chọn 3 bóng hỏng}{Tổng số cách chọn 3 bóng} = \frac{1}{220}$$

    b) Xác suất để cả ba bóng đều không hỏng là: <br>
    - Số cách chọn 3 bóng không hỏng: $C_9^3 = 84$
    $$P(B) = \frac{Số cách chọn 3 bóng không hỏng}{Tổng số cách chọn 3 bóng} = \frac{84}{220}$$

    c) Xác suất để có ít nhất một bóng không hỏng là: <br>
    $$P(C) = 1 - P(\overline{C}) = 1 - P(A) = 1 - \frac{1}{220} = \frac{219}{220} = 0.99$$

    d) Xác suất để chỉ có bóng thứ hai hỏng là: <br>
    - Số cách chọn 2 bóng không hỏng và 1 bóng hỏng: $C_9^2 \times C_3^1 = 27 \times 3 = 81$
    $$P(D) = \frac{Số cách chọn 2 bóng không hỏng và 1 bóng hỏng}{Tổng số cách chọn 3 bóng} = \frac{81}{220} = 0.368$$
